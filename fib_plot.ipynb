```python
%matplotlib inline
from math import sqrt

import numpy as np
import matplotlib.pyplot as plt

# 項数nのフィボナッチ数列を返す関数
def fib(n):
    li = [0,1]
    if n == 1:
        return [0]
    elif n == 2:
        return li
    else:
        for i in range(2,n):
            li.append(li[i-2]+li[i-1])
    return li

# 項数nのフィボナッチ数列の隣り合う数の比の列を返す関数
def ratio(n):
    fibo = fib(n)
    li_rate = [0,0]
    if n == 1:
        return [0]
    elif n == 2:
        return li_rate
    else:
        for i in range(2,n):
            li_rate.append(fibo[i]/fibo[i-1])
    return li_rate

# 項数nのフィボナッチ数列の，隣り合う数の比を可視化
def fib_ratio_draw(n):
    x = np.arange(1,n+1)
    y = ratio(n)
    y2 = np.repeat((1+sqrt(5))/2, n)
    fig, ax = plt.subplots(figsize=(7,6))
    ax.plot(x, y, label="Fibonacci ratio")
    ax.plot(x, y2, linestyle="--", label="Golden ratio")  # 黄金比をplot
    ax.legend()
    ax.set_xlabel("No.")
    ax.set_ylabel("Ratio")
    plt.savefig("fib_plot.png")   # 画像ファイルとして保存
```

```python
fib_ratio_draw(30)  # 項数30のフィボナッチ数列を生成し，隣り合う数の比をplot 
```
